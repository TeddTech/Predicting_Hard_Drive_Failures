There is a reasonable number of classes in the `model` column for each dataframe so these will be handled using one hot encoding. `serial_number` on the other hand has 24955 classes in `dat1` alone. The `serial_number` column in each dataframe will be treated as a categorical factor.

````python
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()

dat1['serial_number'] = encoder.fit_transform(dat1['serial_number'])
dat2['serial_number'] = encoder.fit_transform(dat2['serial_number'])
dat3['serial_number'] = encoder.fit_transform(dat3['serial_number'])

# one hot encoding
dat1 = pd.get_dummies(dat1, columns=['model'])
dat2 = pd.get_dummies(dat2, columns=['model'])
dat3 = pd.get_dummies(dat3, columns=['model'])

# dates need to be converted for model
dat1['timestamp'] = pd.to_datetime(dat1['date'])
dat1['date'] = dat1['timestamp'].astype(np.int64)

dat2['timestamp'] = pd.to_datetime(dat2['date'])
dat2['date'] = dat2['timestamp'].astype(np.int64)

dat3['timestamp'] = pd.to_datetime(dat3['date'])
dat3['date'] = dat3['timestamp'].astype(np.int64)

#order just incase
dat1.sort_values(by=['date'], inplace=True)
dat2.sort_values(by=['date'], inplace=True)
dat3.sort_values(by=['date'], inplace=True)
````

### Hyperparameter Optimization

I will be using `RandomizedSearchCV` to find optimal hyperparameters for the random forest classifer. Each `RandomizedSearchCV` will complete three runs optimizing for one of the three metrics (`recall`, `percision`, `f1`) on each run.

````python
# Utility function to report best scores
def report(results, n_top=3, random_search=random_search, testy=testy):
    for i in range(1, n_top + 1):
        candidates = np.flatnonzero(results['rank_test_score'] == i)
        for candidate in candidates:
            print("Model with rank: {0}".format(i))
            print("Mean validation score: {0:.3f} (std: {1:.3f})".format(
                  results['mean_test_score'][candidate],
                  results['std_test_score'][candidate]))
            print("Parameters: {0}".format(results['params'][candidate]))
            print("")
    probs = random_search.best_estimator_.predict_proba(X_valid)
    # keep probabilities for the positive outcome only
    probs = probs[:, 1]
    # predict class values
    yhat = random_search.best_estimator_.predict(X_valid)
    # calculate precision-recall curve
    precision, recall, thresholds = precision_recall_curve(testy, probs)
    # calculate F1 score
    f1 = f1_score(testy, yhat)
    # calculate precision-recall AUC
    au = auc(recall, precision)
    # calculate average precision score
    ap = average_precision_score(testy, probs)
    print('f1=%.3f auc=%.3f ap=%.3f' % (f1, au, ap))
    # plot no skill
    plt.plot([0, 1], [0.5, 0.5], linestyle='--')
    # plot the precision-recall curve for the model
    plt.plot(recall, precision, marker='.')
    # show the plot
    plt.show()
    print(classification_report(testy, yhat))
    print('Confusion Matrix')
    print(confusion_matrix(testy, yhat))
    print('')
    print('')
````

````python
warnings.filterwarnings('ignore')

model = RandomForestClassifier()
testy = y_valid

from sklearn.metrics import auc

# specify parameters and distributions to sample from

param_dist = {
              "max_depth": sp_randint(1, 7),
              "max_features": sp_randint(4, 15),
              "min_samples_split": sp_randint(3, 20)}
# run randomized search
n_iter_search = 10

for i in ['recall', 'f1', 'precision']:
    
    print('***************** ' +i.upper()+ ' *****************')
    random_search = RandomizedSearchCV(model, param_distributions=param_dist,
                                       n_iter=n_iter_search, cv=3, iid=False, scoring=i)
    start = time()
    random_search.fit(Xtrain, ytrain)
    print("RandomizedSearchCV took %.2f seconds for %d candidates"
          " parameter settings." % ((time() - start), n_iter_search))
    report(results=random_search.cv_results_, random_search=random_search, testy=testy)
````

The performance is horrible. The classifier is unable to pick out any failure cases. Lets see if taking class imbalance into account by using `class_weight` helps. This gives a higher penalty for misclassifying the minority class. I will set the `class_weights` the same as the proportion of imbalance in the data set.

````python
warnings.filterwarnings('ignore')

model = RandomForestClassifier(class_weight={0:.00001, 1:.99999})
testy = y_valid

from sklearn.metrics import auc

# specify parameters and distributions to sample from

param_dist = {
              "max_depth": sp_randint(1, 7),
              "max_features": sp_randint(4, 15),
              "min_samples_split": sp_randint(3, 20)}
# run randomized search
n_iter_search = 10

for i in ['recall', 'f1', 'precision']:
    
    print('***************** ' +i.upper()+ ' *****************')
    random_search = RandomizedSearchCV(model, param_distributions=param_dist,
                                       n_iter=n_iter_search, cv=3, iid=False, scoring=i)
    start = time()
    random_search.fit(Xtrain, ytrain)
    print("RandomizedSearchCV took %.2f seconds for %d candidates"
          " parameter settings." % ((time() - start), n_iter_search))
    report(results=random_search.cv_results_, random_search=random_search, testy=testy)
````

Better but definately not the best. The classifier above is now able to pick observations that are failures. This classifier has a high recall value identifying the majority of failure cases but it is also identifying alot of non failure cases as failures. This high type I error is not practical for this problem. Say we were to tend to all hard drives the classifier identified as failures, alot of money and time will be wasted fixing false positives. Depending on the treatment cost per hard drive even though we will be able to tend to the majority of hard drives before they fail this may not make up for the cost of tending to hard drives that didn't need treatment.

## Resampling

I will now address the class imbalance using sampling techniques. I will use `SMOTETomek` which does both upsampling on the minority class and downsampling on the majority class. This will give the classifier a larger ratio of positive failures to learn.

````python
from imblearn.combine import SMOTETomek

print('Shape of Training set before sampling:', Xtrain.shape)
print('Proportion of failures=1:', np.mean(ytrain))
print('')
col = Xtrain.columns
idx = Xtrain.index

smt = SMOTETomek()
Xtrain, ytrain = smt.fit_sample(Xtrain, ytrain)

Xtrain = pd.DataFrame(Xtrain, columns=col, index=idx)
print('Shape of Training set after sampling:', Xtrain.shape)
print('Proportion of install=1:', np.mean(ytrain))

pd.Series(ytrain).value_counts().plot(kind='bar', title='Count (target)', stacked=True)
plt.title("Sample Set")
plt.ticklabel_format(style='sci', axis='y', scilimits=(0,0))
plt.show()

display(d.failure.value_counts())
display(pd.Series(ytrain).value_counts())
display(y_valid.value_counts())
````

`reallocated_sector` counts the number of sectors moved to the spare area indicating problem with the disk surface or the read/write heads. Because `reallocated_sector` seems to be the best indicator for failing hard drives it is possible that other metrics that track movement of resources may also be good indicators for failing hard drives. Examples of such metrics could include `Reallocation Event Count`, this counts the number of sector remap operations  and `Current Pending Sector Count`, this counts the number of unstable sectors. These pending sectors may be remapped to the spare area.

Including metrics that track physical shape of hard drives may also be good indicators for failing hard drives but the interpretation of these metrics (what values are good or bad) vary with hard drive manufacture and model type. For example, `spin_retry_count` keep track of the retry count of spin start attempts. This may indicate problem with motor, bearings or power supply.
